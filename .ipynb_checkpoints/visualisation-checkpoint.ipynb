## Etude sur le genre du film

Après avoir étudié l'évolution des notes en fonction de l'année, il serait pertinent d'étudier les tendances de préférences pour les films selon leur genres principaux, qui est une variable clé de notre base de données.

Pour ce faire, comme de nombreux films possèdent plusieurs genres, on va devoir split la base de données :

```python
df = split_method.split_elem_col(df[0:100], "genres")
df
```

L'opération devient beaucoup plus longue pour la database complète. Afin de s'épargner de ce temps d'attente, nous avons effectué l'opération uniquement une fois puis avons save la base désirée pour tous les films sous le nom de *data_genres_split.csv* sur le GitHub dans le dossier **data**. Nous l'utiliserons ainsi pour faire les analyses désirées concernant les genres.

On charge ainsi la base *data_genres_split* : 

```python
df_genres_split = pd.read_csv("data/data_genres_split.csv", sep = "\t")
df_genres_split
```

Regardons d'abord quels genres sont les plus représentés :

```python
movies_viz.genres_count(df_genres_split)
```

A partir de "News", on considère que le nombre de films est trop faible pour avoir une réelle signification. On restreint donc l'étude à 22 genres. Traçons les boîtes à moustaches pour ces 22 genres, par deux groupes de 11 pour plus de visibilité :

```python
list_genres_1 = ["Drama", "Comedy", "Romance", "Action", "Crime", "Thriller", "Horror",
         "Adventure", "Mystery", "Family", "Fantasy"]   #Premier groupe

list_genres_2 = ["Sci-Fi", "Musical", "War", "History", "Western", "Animation", "Music",
         "Biography", "Adult", "Sport", "Film-Noir"]   #Deuxième groupe
```

```python
movies_viz.genres_means(df_genres_split, list_genres_1)   #Note moyenne des 11 genres les plus représentés
```

```python
movies_viz.genres_boxplot(df_genres_split, list_genres_1)
```

```python
movies_viz.genres_means(df_genres_split, list_genres_2)   #Note moyenne du deuxième groupe
```

```python
movies_viz.genres_boxplot(df_genres_split, list_genres_2)
```

Les dramas, les films de guerre, les films historiques et biographiques semblent les plus qualitatifs. Mais les dramas sont présents en plus grands nombres, donc leur score est plus significatif.
Certains genres tels que la Science-Fiction et les films d'horreur semblent proposer le meilleur comme le pire.
